# Ann Arbor, Michigan Record Temperatures from 2004-2015

The data for this project comes from a subset of The National Centers for Environmental Information (NCEI) [Daily Global Historical Climatology Network](https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/readme.txt) (GHCN-Daily). The GHCN-Daily is comprised of daily climate records from thousands of land surface stations across the globe.

Each row in the datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)
    * TMIN : Minimum temperature (tenths of degrees C)
* **value** : data value for element (tenths of degrees C)

For this project, I plan to:

4. Make the visual nice! Leverage principles from the first module in this course when developing your solution. Consider issues such as legends, labels, and chart junk.

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('tyler/AnnArborTemps.csv')

df.sort_values(['Date'], inplace=True)
df['Year'] = df['Date'].apply(lambda x: (x[:4]))
df['Month-Day'] = df['Date'].apply(lambda x: (x[5:]))
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df = df[df['Month-Day'] != '02-29']
df15 = df[df['Year'] == '2015']
df = df[df['Year'] != '2015']
df['Year'].unique()
```

```python
tmin = df[df['Element'] == 'TMIN'].groupby('Month-Day').apply(np.min)
tmax = df[df['Element'] == 'TMAX'].groupby('Month-Day').apply(np.max)
t15min = df15[df15['Element'] == 'TMIN'].groupby('Month-Day').apply(np.min)
t15max = df15[df15['Element'] == 'TMAX'].groupby('Month-Day').apply(np.max)

tmin.head()
```

```python
tmax.head()
```

```python
minRecordBreak = np.where(tmin['Data_Value'] > t15min['Data_Value'])[0]
maxRecordBreak = np.where(tmax['Data_Value'] < t15max['Data_Value'])[0]
```

```python
tmin = tmin['Data_Value']
tmax = tmax['Data_Value']
t15min = t15min['Data_Value']
t15max = t15max['Data_Value']
```

```python
tmin.head()
```

```python
tmax.head()
```

```python
print(minRecordBreak, maxRecordBreak)
```

```python
plt.figure(figsize=(12,6))
plt.plot(tmin.values, color='m', label='Record Low')
plt.plot(tmax.values, color='g', label='Record High')
plt.scatter(minRecordBreak, t15min.iloc[minRecordBreak], s=10, color='k', label='Min Temp Record\nBreak in 2015')
plt.scatter(maxRecordBreak, t15max.iloc[maxRecordBreak], s=10, color='b', label='Max Temp Record\nBreak in 2015')
plt.gca().fill_between(range(len(tmin)), tmin.values, tmax.values, facecolor='y', alpha=.15)
plt.xticks(range(0, len(tmin), 30), tmin.index[range(0, len(tmin), 30)], rotation=45)
plt.xlabel('Day of the Year')
plt.ylabel('Temperature (tenths of degrees C)')
plt.title('Record Highs and Lows by Date for Detroit Michigan')
plt.legend(frameon=False)
plt.show()
```
